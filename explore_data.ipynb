```python
def load_index(which_index: str = "NR"):
    
    result = {}
    
    # Load the index file, which is of JSON format
    with open(core_database_index_dir + f"DB_{which_index}.json") as f:
        index = json.load(f)
    
    data = index["data"]
    
    # Go over all data, which is a list of dictionaries, and extract the information, with key being database_key
    for d in data:
        result[d["database_key"]] = d
    
    return result
```
